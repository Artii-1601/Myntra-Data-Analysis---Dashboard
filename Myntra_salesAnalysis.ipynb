```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
#checking sheetsname in the excel file
pd.ExcelFile("/content/Myntra dataset.xlsx").sheet_names
```

##loading data

```
#loading data
products=pd.read_excel("/content/Myntra dataset.xlsx",sheet_name='dim_products')
products.info()
products
```

```
customers=pd.read_excel("/content/Myntra dataset.xlsx",sheet_name='dim_customers')
customers.info()
customers
```

```
orders=pd.read_excel("/content/Myntra dataset.xlsx",sheet_name='fact_orders')
orders.info()
orders
```

##Data inspection

```
#identify duplicate rows in a DataFrame.

print(products.duplicated().sum())
print(customers.duplicated().sum())
print(orders.duplicated().sum())
```

```
products.drop_duplicates()
```

##Data transformation

```
###created month column through date column in orders

orders["Month"]=orders["Date"].dt.strftime("%B")
#dt.strftime("%B") is used to datetime object in pandas to return the full name of the month.
orders
```

```
# total price
orders["Total Price"]=orders["Original Price"]-(orders["Original Price"]*orders["Discount%"])
orders
```

#mergeing the table

```
df = pd.merge(left = orders, right= products, on = "Product ID", how = "inner" )
df
```

```
df["Total Price"].sum()
```

```
df1=df.merge(customers,on="Customer ID")
df1
```

# **Data Analysis**

##calcualting Revenue

```
df1['Revenue']=df1['Original Price'] * (1-df1['Discount%'])
df1
```

##total revenue

```
#total revenue

total_revenue=df1['Revenue'].sum()
total_revenue
```

## total orders

```
#total orders
total_orders=df1['Order ID'].nunique()
total_orders
```

##Top 5 best-selling Products and Brands(by revenue)


```
# top_prodcuts=df1.groupby('Product Name').agg({'Revenue':'sum'})
# top_prodcuts.nlargest(5,'Revenue')
top_5_prodcuts=df1.groupby('Product Name')['Revenue'].sum().nlargest(5)
top_5_prodcuts
```

```
plt.figure(figsize=(10,4))
sns.barplot(x=top_5_prodcuts.index,y=top_5_prodcuts.values,hue=top_5_prodcuts.index,palette='summer',legend=False)
plt.xticks(rotation=45)
```

```
top_5_Brands=df1.groupby('Brand Name')['Revenue'].sum().nlargest(5)
top_5_Brands
```

##sales by state

top buying state

```
sales_by_state=df1.groupby('State')['Revenue'].sum().nlargest(5)
sales_by_state
```

##montly sales trend

```
monthly_sales=df1.groupby('Month')['Revenue'].sum()
# monthly_sales.sort_values(ascending=False)
```

```
plt.figure(figsize=(10,4))

sns.barplot(x = monthly_sales.index, y = monthly_sales.values, order = ["January", "February", "March",'April',"May","June","July","August","September","October","November", "December"],hue = monthly_sales.index, palette = "GnBu")
plt.xticks(rotation = 45)
plt

# Convert the Series to a DataFrame for countplot
monthly_sales_df = monthly_sales.reset_index()

#showing the values
a = sns.countplot(x='Month', data=monthly_sales_df, order=["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"], hue='Month', palette="GnBu") # Updated countplot parameters
for i in a.containers:
  a.bar_label(i)
```

##6.Impact of discount on sales

```
discount_Sales=df1.groupby('Discount%')['Revenue'].sum()
discount_Sales.nlargest(5)
```
